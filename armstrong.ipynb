```python
n=int(input("enter a number"))
sum=0
original=n
while n>0:
    i=n%10
    sum=sum+i*i*i
    n=n//10
if sum==original:
    print("armstrong number")
else:
    print("not an armstrong number")
```

```python
u=int(input("enter upper limit:"))
l=int(input("enter lower limit:"))

for i in range (l,u+1):
    temp=i
    sum=0
    while temp>0:
        d=temp%10
        sum=sum+d*d*d
        temp=temp//10
    if sum==i:
        print(i)
    else:
        continue
        
```
